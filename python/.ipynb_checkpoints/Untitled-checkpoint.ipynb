```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load the grayscale image
image = cv2.imread('transonic.jpeg', cv2.IMREAD_GRAYSCALE)

# Step 2: Apply Gaussian Blur to reduce noise
blurred = cv2.GaussianBlur(image, (5, 5), 0)

# Step 3: Crop the bottom region (adjust 0.8 as needed)
height, width = blurred.shape
bottom_region = blurred[int(height * 0.8):, :]  # Bottom 20% of the image

# Step 4: Apply edge detection (Canny)
edges = cv2.Canny(bottom_region, threshold1=50, threshold2=150)

# Step 5: Extract the lowest line
height, width = edges.shape
max_height = np.zeros(width)

# Find the lowest edge pixel for each column
for h in range(height - 1, -1, -1):  # Iterate from bottom to top
    for w in range(width):
        if edges[h, w] != 0 and max_height[w] == 0:  # First non-zero pixel
            max_height[w] = h

# Step 6: Create a blank mask to draw the lowest line
lowest_line_mask = np.zeros_like(edges)

# Draw the lowest line
for w, h in enumerate(max_height):
    if h > 0:  # Only draw for valid heights
        lowest_line_mask[int(h), w] = 255

# Step 7: Display results
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')

plt.subplot(1, 3, 2)
plt.title("Edges in Bottom Region")
plt.imshow(edges, cmap='gray')

plt.subplot(1, 3, 3)
plt.title("Lowest Line")
plt.imshow(lowest_line_mask, cmap='gray')

plt.tight_layout()
plt.savefig("lowest_line.png", dpi=300, bbox_inches='tight', transparent=True)
plt.show()
```

```python
import numpy as np

# Parameters for the mesh
num_layers = 10           # Number of vertical layers
vertical_spacing = 5       # Spacing between layers (in pixels or model units)
width = len(max_height)    # Number of columns in the profile

# Initialize a list to store the mesh points
mesh_points = []

# Generate the mesh points
for layer in range(num_layers):
    for x, base_y in enumerate(max_height):
        if base_y > 0:  # Valid profile point
            y = base_y + layer * vertical_spacing
            mesh_points.append((x, y, layer))  # Append (x, y, z) coordinates

# Save the mesh to a file (e.g., .txt or .msh)
with open("geomesh.txt", "w") as file:
    file.write("X Y Z\n")  # Header
    for point in mesh_points:
        file.write(f"{point[0]} {point[1]} {point[2]}\n")

print(f"Mesh file 'geomesh.txt' created with {len(mesh_points)} points.")
```

```python
import matplotlib.pyplot as plt

# Extract x, y, and z coordinates for visualization
x_coords = [p[0] for p in mesh_points]
y_coords = [p[1] for p in mesh_points]
z_coords = [p[2] for p in mesh_points]

plt.figure(figsize=(10, 5))
plt.scatter(x_coords, y_coords, c=z_coords, cmap='viridis', s=10)
plt.title("Generated Geomesh")
plt.xlabel("X")
plt.ylabel("Y")
plt.colorbar(label="Layer (Z)")
plt.show()
```
